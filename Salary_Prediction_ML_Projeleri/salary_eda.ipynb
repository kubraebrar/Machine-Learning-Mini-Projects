```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi oku
df = pd.read_csv('Salary_dataset.csv', index_col=0)

# İlk 5 satıra bakalım
df.head()
```

```python
#shape bize satır sayısı verir 
satir_sayisi = df.shape[0]
print(f"Veri setindeki toplam satır sayısı: {satir_sayisi}")
```

```python
#outlier bulmak için Boxplot kullanıcaz, Eğer grafiğin alt veya üst çizgilerinin dışında kalan tek tük noktalar varsa, onlar outlier'dır.
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Salary'], color='skyblue')
plt.title('Maaş Dağılımı (Outlier Kontrolü)')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Salary'], bins=10, kde=True, color='purple')
plt.title('Maaşların Frekans Dağılımı')
plt.xlabel('Maaş')
plt.ylabel('Kişi Sayısı')
plt.show()
```

```python
korelasyon = df.corr()
print("Korelasyon Matrisi:\n", korelasyon)

# Daha şık görmek için Isı Haritası (Heatmap)
sns.heatmap(korelasyon, annot=True, cmap='coolwarm')
plt.title('Korelasyon Isı Haritası')
plt.show()
```

```python
yearsExperience = df[['YearsExperience']]
salary = df['Salary']

from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(yearsExperience,salary,test_size=00.33, random_state=0)

#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

tahmin = lr.predict(x_test)

x_train = x_train.sort_index()
y_train = y_train.sort_index()

plt.plot(x_train,y_train)
plt.plot(x_test, lr.predict(x_test))


```

```python
tahmin_7_yil = lr.predict([[7]])
print(f"7 yıllık deneyimi olan birinin tahmini maaşı: {tahmin_7_yil[0]:.2f}")
```
