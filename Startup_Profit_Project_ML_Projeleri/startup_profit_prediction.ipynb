```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('50_Startups.csv')

# Sadece sayısal (matematiksel) verileri içeren kolonları listele
sayisal_kolonlar = df.select_dtypes(include=['number']).columns.tolist()
print("Sayısal Featurelar (Numerical):")
print(sayisal_kolonlar)

# Sadece metin/yazı (kategorik) içeren kolonları listele
kategorik_kolonlar = df.select_dtypes(include=['object']).columns.tolist()
print("Kategorik Featurelar (Categorical):")
print(kategorik_kolonlar)


```

```python
#Missing value var mı bak ? 
print((df == 0).sum())
```

```python
df_encoded = pd.get_dummies(df,columns=['State'],drop_first=True)
df_encoded.head()
```

```python
#Train/Test Split
from sklearn.model_selection import train_test_split

X = df_encoded.drop('Profit', axis=1)
y = df_encoded['Profit']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#multiple regression 
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

#Prediction
y_pred = model.predict(X_test)

final = pd.DataFrame({'Real Profit (y_test)': y_test, 'Model Prediction (y_pred)': y_pred})
print(final)
```

```python
# Backward Elimination
import pandas as pd
import statsmodels.api as sm 
import numpy as np

X = df_encoded.drop('Profit', axis=1)
y = df_encoded['Profit']

X = np.append(arr = np.ones((50,1)).astype(int),values= X ,axis=1)

X_l = X[:,[0,1,2,3,4,5]]
X_l =np.array(X_l,dtype=float)

regressor_OLS = sm.OLS(endog =y, exog = X_l).fit()
print(regressor_OLS.summary())
```

```python
X_l = X[:,[0,1,2,3,4]]
X_l =np.array(X_l,dtype=float)

regressor_OLS = sm.OLS(endog =y, exog = X_l).fit()
print(regressor_OLS.summary())
```
